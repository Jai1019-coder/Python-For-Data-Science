```python
import matplotlib.pyplot as plt
```

```python
x = [i for i in range(1,21)]
y = [j**2 for j in x]
```

```python
plt.plot(x,y)
```

```python
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Basic Line Plot')
plt.plot(x,y)
```

```python
x = [i for i in range(1,6)]
y = [j**2 for j in x]
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Basic Line Plot')
plt.plot(x,y, color = 'red', linestyle='--', marker = 'o', linewidth = 0.5, markersize = 0.75)
plt.grid(True)
```

### Mutiple Plots

```python
x = [1,2,3,4,5,]
y1 = [1,4,9,16,25]
y2 = [1,2,3,4,5]
```

```python
plt.figure(figsize=(100,100))

plt.subplot(2,3,1)
plt.plot(x,y1,color="green")
plt.title("Plot 1")

plt.subplot(2,2,2)
plt.plot(y1,x,color="green")
plt.title("Plot 2")

plt.subplot(2,2,3)
plt.plot(x,y2,color="blue")
plt.title("Plot 3")

plt.subplot(2,2,4)
plt.plot(x,y2,color="red")
plt.title("Plot 4")
```

# Sales Data Visualisation

```python
import pandas as pd
sales_data_df = pd.read_csv('sales_data.csv')
sales_data_df.head(5)
```

### Total Sales by Product Category

```python
total_sales_by_category = sales_data_df.groupby('Product Category')['Total Revenue'].sum()
total_sales_by_category.plot(kind = 'bar', color = 'teal', ylabel='Total Revenue')
```

### Sales trend by Date

```python
sales_trend = sales_data_df.groupby('Date')['Total Revenue'].sum().reset_index()
plt.plot(sales_trend['Date'],sales_trend['Total Revenue'])
```

### Percentage Sales by Region

```python
sales_by_region = sales_data_df.groupby('Region')['Total Revenue'].sum().reset_index()
plt.pie(sales_by_region['Total Revenue'],labels=sales_by_region['Region'],colors=['gold','yellowgreen',
        'lightcoral'], autopct="%1.2f%%")
```
