# Data Manipulation and Analysis

```python
import pandas as pd
df = pd.read_csv('data.csv')
df.tail()
```

```python
df.dtypes
```

```python
df.describe()
```

```python
print(df.isnull().any(axis = 1).sum())
print(df.isnull().sum())
```

```python
df_filled = df['Value'].fillna(df['Value'].mean(), inplace=True)
print(df)
```

```python
df.rename(columns={'Date':'Sale Date'}, inplace=True)
df.head()
```

```python
df['Value']=df['Value'].astype(int)
print(df)
```

```python
df.rename(columns={'Old Value':'Old_Value'},inplace=True)
print(df)
```

```python
df['New_Value'] = df['Old_Value'].apply(lambda x : x+x*0.5)
df
```

```python
df['Sales'] = df['Sales'].fillna(df['Sales'].mean())
```

```python
old_grouped_mean = df.groupby('Product')['Old_Value'].mean()
print(old_grouped_mean)
new_grouped_mean = df.groupby('Product')['New_Value'].mean()
print(new_grouped_mean)
```
